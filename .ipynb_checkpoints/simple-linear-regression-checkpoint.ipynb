## Ground Cricket Chirps

In _The Song of Insects_ (1948) by George W. Pierce, Pierce mechanically measured the frequency (the number of wing vibrations per second) of chirps (or pulses of sound) made by a striped ground cricket, at various ground temperatures.  Since crickets are ectotherms (cold-blooded), the rate of their physiological processes and their overall metabolism are influenced by temperature.  Consequently, there is reason to believe that temperature would have a profound effect on aspects of their behavior, such as chirp frequency.

In general, it was found that crickets did not sing at temperatures colder than 60&deg; F or warmer than 100&deg; F.

```python
ground_cricket_data = {"Chirps/Second": [20.0, 16.0, 19.8, 18.4, 17.1, 15.5, 14.7,
                                         15.7, 15.4, 16.3, 15.0, 17.2, 16.0, 17.0,
                                         14.4],
                       "Ground Temperature": [88.6, 71.6, 93.3, 84.3, 80.6, 75.2, 69.7,
                                              71.6, 69.4, 83.3, 79.6, 82.6, 80.6, 83.5,
                                              76.3]}
df = pd.DataFrame(ground_cricket_data)
```

### Tasks

1. Find the linear regression equation for this data.
2. Chart the original data and the equation on the chart.
3. Find the equation's $R^2$ score (use the `.score` method) to determine whether the equation is a good fit for this data. (0.8 and greater is considered a strong correlation.)
4. Extrapolate data:  If the ground temperature reached 95&deg; F, then at what approximate rate would you expect the crickets to be chirping?
5. Interpolate data:  With a listening device, you discovered that on a particular morning the crickets were chirping at a rate of 18 chirps per second.  What was the approximate ground temperature that morning?

```python
X = df["Chirps/Second"].reshape(-1, 1)
y = df["Ground Temperature"].reshape(-1, 1)

lr = linear_model.LinearRegression()
lr.fit(X, y)

lr2 = linear_model.LinearRegression()
lr2.fit(y, X)

print('Intercept:', lr.intercept_)
print('Coefficient:', lr.coef_)
print('R2|Score:', lr.score(X, y))
```

# Seems that this equation isn't a great fit for the data with an R2 score of .69

```python
plt.scatter(X, y, color='black')
plt.plot(X, lr.predict(X), color='b')
plt.show()
```

```python
print("Temp at 18 Chirps/sec:", lr.predict(18))
print("Chirps/sec at 95 degrees:", lr2.predict(95))
```

## Salary Discrimination

The file `salary.txt` contains data for 52 tenure-track professors at a small Midwestern college. This data was used in legal proceedings in the 1980s about discrimination against women in salary.

The data in the file, by column:

1. Sex. 1 for female, 0 for male.
2. Rank. 1 for assistant professor, 2 for associate professor, 3 for full professor.
3. Year. Number of years in current rank.
4. Degree. Highest degree. 1 for doctorate, 0 for master's.
5. YSdeg. Years since highest degree was earned.
6. Salary. Salary/year in dollars.

### Tasks

1. Find the linear regression equation for this data using columns 1-5 to column 6.
2. Find the selection of columns with the best $R^2$ score.
3. Report whether sex is a factor in salary. Support your argument with graph(s) if appropriate.

```python
df = pd.read_fwf("salary.txt", header=None, 
                 names=["Sex", "Rank", "Year", "Degree", "YSdeg", "Salary"])
```

```python
sex = df['Sex'].reshape(-1, 1)
rank = df['Rank'].reshape(-1, 1)
year = df['Year'].reshape(-1, 1)
degree = df['Degree'].reshape(-1, 1)
YSdeg = df['YSdeg'].reshape(-1, 1)
salary = df['Salary'].reshape(-1, 1)

sexlr = linear_model.LinearRegression()
sexlr.fit(sex, salary)

ranklr = linear_model.LinearRegression()
ranklr.fit(rank, salary)

yearlr = linear_model.LinearRegression()
yearlr.fit(year, salary)

degreelr = linear_model.LinearRegression()
degreelr.fit(degree, salary)

YSdeglr = linear_model.LinearRegression()
YSdeglr.fit(YSdeg, salary)

plt.figure(figsize = (12, 12))

plt.subplot(321)
plt.title('Sex')
plt.scatter(sex, salary)
plt.plot(sex, sexlr.predict(sex))

plt.subplot(322)
plt.title('Rank')
plt.scatter(rank, salary)
plt.plot(rank, ranklr.predict(rank))

plt.subplot(323)
plt.title('Year')
plt.scatter(year, salary)
plt.plot(year, yearlr.predict(year))

plt.subplot(324)
plt.title('Degree')
plt.scatter(degree, salary)
plt.plot(degree, degreelr.predict(degree))

plt.subplot(325)
plt.title('YSdeg')
plt.scatter(YSdeg, salary)
plt.plot(YSdeg, YSdeglr.predict(YSdeg))

plt.show()
```

```python
score = pd.DataFrame(data={'R2 Score':[sexlr.score(sex, salary), ranklr.score(rank, salary),yearlr.score(year, salary),
                           degreelr.score(degree, salary), YSdeglr.score(YSdeg, salary)]},
                     index=["Sex", "Rank", "Year", "Degree", "YSdeg"])
score.sort_values(by='R2 Score', ascending=False)
# print('Sex R2|Score:', sexlr.score(sex, salary))
# print('Rank R2|Score:', ranklr.score(rank, salary))
# print('Year R2|Score:', yearlr.score(year, salary))
# print('Degree R2|Score:', degreelr.score(degree, salary))
# print('YSdeg R2|Score:', YSdeglr.score(YSdeg, salary))
```

# Rank has a very high R2 score and is a good illustrator of a healthy growth in salary as they grow in their careers. Sex is definitely a factor in salaries. It's fairly obvious to see the difference in salaries between men and women and none of the other factors would have such an effect.
